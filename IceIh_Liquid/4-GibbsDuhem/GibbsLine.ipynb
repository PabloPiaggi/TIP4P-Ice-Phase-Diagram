# Gibbs-Duhem integration

```python
import numpy as np
import glob, os
import matplotlib.pylab as plt
```

```python
PT=[]
for fold in glob.glob('Gibbs_Duhem_Simulations/*_*/')+glob.glob('Gibbs_Duhem_Simulations_Neg/*_*/'):
    print(fold)

    fold=fold.split('/')[-2]
   # print(fold)
    PT.append([float(fold.split('_')[1]),float(fold.split('_')[2].strip('/'))])
PT=np.array(PT)
PT=PT[np.argsort(PT[:,1])]
```

```python
plt.plot(PT[::4,0],PT[::4,1])
plt.scatter(PT[:,0],PT[:,1])
```

```python
np.savetxt('T_P.dat',PT)
```
