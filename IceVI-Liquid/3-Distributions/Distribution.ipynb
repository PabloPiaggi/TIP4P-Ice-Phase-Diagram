# Distribution of order parameter

```python
import numpy as np, glob
import matplotlib.pylab as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
```

### Overlap between distributions

```python
data=np.loadtxt('results.txt')

plt.figure(figsize=(3.4,2))
plt.plot(data[:,0],data[:,1])
plt.xlabel(r'$\sigma$')
plt.ylabel('Overlap')
plt.tight_layout()
```

```python
sigma_min=data[data[:,1]==min(data[:,1])][0][0]
print('Minimum overlap for sigma = {}'.format(sigma_min))
```

### Distributions

```python
hist_liquid = np.loadtxt('Liquid/histo')
hist_IceVI = np.loadtxt('IceVI/histo')
```

```python
plt.figure(figsize=(3.4,2))
plt.plot(hist_IceVI[:,0],hist_IceVI[:,1])
plt.plot(hist_liquid[:,0],hist_liquid[:,1])
plt.xlim([0.3,1.3])
plt.annotate(r'$\sigma={}$'.format(sigma_min),(0.4,10))

max_ice=hist_IceVI[hist_IceVI[:,1]==hist_IceVI[:,1].max()][0][0]
max_liquid=hist_liquid[hist_liquid[:,1]==hist_liquid[:,1].max()][0][0]


diff_val = np.abs(hist_liquid[:,1]-hist_IceVI[:,1])
diff_bin = hist_liquid[:,0]
diff_val = diff_val[(diff_bin>max_liquid) * (diff_bin<max_ice)]
diff_bin = diff_bin[(diff_bin>max_liquid) * (diff_bin<max_ice)]

min_order= diff_bin[diff_val.min()==diff_val]
plt.axvline(x=min_order, color='k', linestyle='--')
plt.axvline(x=max_liquid, color='k', linestyle='--')
plt.axvline(x=max_ice, color='k', linestyle='--')
```

```python
print('Liquid max:', max_liquid)
```
